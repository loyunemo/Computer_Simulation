```python
import numpy as np
def read_dat_file(file_path):
    try:
        with open(file_path, 'r') as file:
            # 读取文件中的每一行
            lines = file.readlines()

            # 初始化空列表，用于存储 x、y、z 坐标
            Coord = []

            # 解析每一行的数据
            for line in lines:
                # 分割每一行的数据
                data = line.strip().split(',')
                # 确保每行有三个数据
                if len(data) == 3:
                    # 将数据转换为浮点数并添加到相应的列表中
                    Coord.append([float(data[0]),float(data[1]),float(data[2])])
                else:
                    print(f"跳过不完整的行: {line}")

            # 返回三个坐标列表
            return Coord
    except Exception as e:
        print(f"读取文件时发生错误: {e}")
        return None


file_path = 'ex1.dat'  # 将 'your_file.dat' 替换为你的dat文件路径
data = np.array(read_dat_file(file_path))
print(Coord)
```

```python
import numpy as np
import matplotlib.pyplot as plt


x, y, z = data[:, 0], data[:, 1], data[:, 2]

# 设置迭代次数和学习率
num_iterations = 1000
learning_rate = 0.01

# 初始化参数 a, b, c
a, b, c = 0, 0, 0

# 梯度下降迭代
for i in range(num_iterations):
    # 计算模型预测值
    predictions = a + b * x + c * y
    
    # 计算残差
    residuals = predictions - z
    
    # 更新参数
    a = a - learning_rate * np.sum(residuals) / len(z)
    b = b - learning_rate * np.sum(residuals * x) / len(z)
    c = c - learning_rate * np.sum(residuals * y) / len(z)

# 打印最终拟合结果
print(f'a: {a}, b: {b}, c: {c}')

# 绘制散点图和拟合平面
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='r', marker='o')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# 生成拟合平面上的点
fit_plane = a + b * x + c * y
X, Y = np.meshgrid(x, y)
ax.plot_surface(X,
                Y,
                a + b * X + c * Y,
                color='b',
                alpha=0.6
               )

plt.show()
```

```python
def batch_gradient_descent(x, y, alpha, num_iterations):
    for i in range(num_iterations):
        dx = 2 * x
        dy = 2 * y
        x = x - alpha * dx
        y = y - alpha * dy
    return (x, y)
(minx,miny)=batch_gradient_descent(7,6,0.01,10000)
print("minx:",minx,"miny:",miny)
```

```python
def stochastic_gradient_descent(x, y, alpha, num_iterations):
    for i in range(num_iterations):
        dx = 2 * x
        dy = 2 * y
        x = x - alpha * dx
        y = y - alpha * dy
    return (x, y)
```
