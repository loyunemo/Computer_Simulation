Gibbs抽样算法模拟计算圆周率$\pi$

```python
class point():
    def __init__(self) -> None:
        self.X=0
        self.Y=0
    def Xaxis(self, X):
        self.X=X
    def Yaxis(self, Y):
        self.Y=Y
def NextIter():
    pointNext=point()
    pointNext.Xaxis(2*random.random()-1)
    pointNext.Yaxis(2*random.random()-1)
    return pointNext
def Addlist(Lst,addpoint):
    Lst.append(addpoint)
```

```python
pointLst=[]
pointzero=point()
pointzero.Xaxis(2*random.random()-1)
pointzero.Yaxis(2*random.random()-1)
pointLst.append(pointzero)

def simulateG(times):
    for t in range(times-1):
        Addlist(pointLst,NextIter())
simulateG(10000)
length=len(pointLst)
count=0
for t in pointLst:
    if (t.X*t.X+t.Y*t.Y)<=1:
        count+=1
    else:
        pass
pi=count/length*4
print(pi)
```

蓄水池抽样算法

```python
def reservoir(Targetlength,Data):
    lst=[]
    for t in range(Targetlength):
        lst.append(Data[t])
    for t in range(Targetlength,len(Data)):
        d = random.randint(0,t-1)
        if d < Targetlength:
            lst[d]=Data[t]
    return lst
def randomInit():
    length=random.randint(15,37268)
    lstdata=[]
    
    for t in range(length):
        lstdata.append(random.randint(0,37267))
    return lstdata
data=randomInit()
target=reservoir(15,data)
print(target)
```
