```python
import numpy as np
import matplotlib.pyplot as plt

# 创建数据
mu1, sigma1 = 0, 1  # 正态分布1的均值和标准差
mu2, sigma2 = 2, 1  # 正态分布2的均值和标准差
x = np.linspace(-3, 5, 1000)  # 生成x轴数据点
def custom_probability_distribution(x, mean, std_dev, scale):
    # 使用高斯分布函数创建一个基本形状
    gaussian = 1 / (std_dev * np.sqrt(2 * np.pi)) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)
    
    # 进行y轴轴对称变换
    custom_distribution = scale * (1 - np.abs(2 * gaussian - 1))
    
    return custom_distribution

y1 = 1 / (sigma1 * np.sqrt(2 * np.pi)) * np.exp(-(x - mu1) ** 2 / (2 * sigma1 ** 2))
y2 = custom_probability_distribution(x,0,1,1)

# 绘制正态分布函数
plt.plot(x, y1, label='N(0, 1)')
plt.plot(x, y2, label='G(0,1,1)')

# 添加标题和标签
plt.title('Multiple Normal Distributions')
plt.xlabel('X-axis')
plt.ylabel('Probability Density')
plt.legend()

# 显示图形
plt.show()
```
