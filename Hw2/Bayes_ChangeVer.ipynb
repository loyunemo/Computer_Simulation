```python
import numpy as np
import re
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

## 原理

### 计算先验概率、后验概率
$$P(C|X)=\frac{P(X|C)P(C)}{P(X)}$$

根据上述公式我们可以知道，样本：训练/测试训练集，假定(H/C):属于垃圾邮件/不属于垃圾邮件

X:特征：所有词为词向量，统计词频，即每个单词在给定标签的出现概率，

P(X|C):计算词频出现概率

P(C):训练集垃圾邮件出现概率
因为P(X)都是给定训练集,所以在计算的时候可以暂时搁置，两个都是同分母，比分子就好，最后的效果就是垃圾邮件/不属于垃圾邮件的概率相加和等于1计算出P(X),从而算出后验概率，如果后验概率大于训练集中垃圾邮件出现概率，认为是垃圾邮件，否则就不是（实际贝叶斯直接调库就好……）

```python
#数据集制作
def textParse(bigString): 
    listOfTokens = re.split('\W+', bigString)
    print(listOfTokens)
    return [tok.lower() for tok in listOfTokens if len(tok) > 2]#全部转换成小写字母
```

输入：词频向量
输出：垃圾邮件 非垃圾邮件标签（0/1）

```python
docList = []
classList = []
#制作数据集，并提取词频向量
for i in range(1, 26): 
    wordList = textParse(open('./email/spam/%d.txt' % i, 'r').read()) 
    docList.append(wordList)#按照顺序来的，一篇邮件一个List
    classList.append(1)#文章标签
    wordList = textParse(open('./email/ham/%d.txt' % i, 'r').read()) 
    docList.append(wordList)
    classList.append(0)
print(docList)
```

```python
def createVocabList(dataSet):#找出所有的单词集合
    vocabSet = set([]) 
    for document in dataSet:
        vocabSet = vocabSet | set(document) 
    return list(vocabSet)
```

```python
vocablist=createVocabList(docList)
DataSetX=[]
DataSetY=classList
#制作训练集#制作统计词频向量
for StringDoc in docList:
    returnVec = [1] * len(vocablist)
    for word in StringDoc:
        if word in vocablist:
            returnVec[vocablist.index(word)] += 1
    DataSetX.append(returnVec)
    print(DataSetX)    
X_train, X_test, y_train, y_test = train_test_split(DataSetX, DataSetY, test_size=0.2, random_state=0)


```

```python
clf = MultinomialNB(force_alpha=True)#贝叶斯分类

clf.fit(X_train,y_train)#计算先验概率
t=list(clf.predict(X_test))
print(t)
print(y_test)
ErrorCount=0
Num=0
IndexErr=[]
for Out in t:
    
    if Out!=y_test[Num]:
        ErrorCount+=1
        IndexErr.append(Num)
    Num+=1
Indexdoc=[]
for Ind in IndexErr:
    Indexdoc.append(DataSetX.index(X_test[Ind]))    
print("Error Probability:%d %%"%(ErrorCount/(len(y_test)*1.0)*100.0))
for Ob in Indexdoc:#打印错误词标签
    print(docList[Ob])
```
